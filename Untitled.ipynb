```python
#to inspect which features are available
import pandas as pd

# Load your saved CSV
df = pd.read_csv("output_filename.csv")

# Identify all ICP40 columns
icp40_columns = [col for col in df.columns if col.endswith('_ICP40')]

# Check how many non-null values each ICP40 column has
non_null_counts = df[icp40_columns].notnull().sum().sort_values(ascending=False)

# Display columns with good data coverage (e.g., more than 70,000 rows)
good_columns = non_null_counts[non_null_counts > 70000]
print(good_columns)
print(f"Number of usable ICP40 features: {len(good_columns)}")
```

```python

# Full End-to-End Pipeline Code Using Merged 

import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load your two CSVs
df_features = pd.read_csv("./data/output_filename.csv")         # from .dbf (contains _ICP40 values and LABNO)
df_locations = pd.read_csv("./data/ngs_shapefile.csv")         # from .shp (contains LABNO and geometry)

# Merge on LABNO
df_merged = pd.merge(df_features, df_locations[['LABNO', 'geometry']], on='LABNO')

# Extract latitude and longitude from geometry column
df_merged[['LONGITUDE', 'LATITUDE']] = df_merged['geometry'].str.extract(r'POINT \((-?\d+\.\d+) (-?\d+\.\d+)\)').astype(float)

# Select well-populated ICP40 columns manually (35 columns you provided)
selected_icp40 = [
    'AL_ICP40', 'CA_ICP40', 'FE_ICP40', 'K_ICP40', 'NA_ICP40', 'TI_ICP40', 'AG_ICP40', 'CO_ICP40', 'CE_ICP40',
    'CD_ICP40', 'BI_ICP40', 'BE_ICP40', 'LI_ICP40', 'LA_ICP40', 'MO_ICP40', 'TH_ICP40', 'SN_ICP40', 'CR_ICP40',
    'CU_ICP40', 'GA_ICP40', 'PB_ICP40', 'SC_ICP40', 'NI_ICP40', 'Y_ICP40', 'V_ICP40', 'U_ICP40', 'ZN_ICP40',
    'BA_ICP40', 'MG_ICP40', 'SR_ICP40', 'MN_ICP40', 'P_ICP40', 'AU_ICP40', 'AS_ICP40', 'NB_ICP40'
]

# Drop rows with NaNs in selected columns or missing lat/lon
df_model = df_merged.dropna(subset=selected_icp40 + ['LATITUDE', 'LONGITUDE'])

# Create unsupervised location clusters using lat/lon
kmeans = KMeans(n_clusters=10, random_state=42)
df_model['region_label'] = kmeans.fit_predict(df_model[['LATITUDE', 'LONGITUDE']])

# Separate features and labels
X = df_model[selected_icp40]
y = df_model['region_label']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Predict and evaluate
y_pred = clf.predict(X_test)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(10, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted Region")
plt.ylabel("True Region")
plt.title("Confusion Matrix: Predicted vs Actual Regions")
plt.tight_layout()
plt.show()

report
```

```python

df_chem = pd.read_csv("./data/output_filename.csv")       # LABNO + ICP40
df_form = pd.read_csv("./data/geochem.csv")            # LABNO + FORMATION (or similar)
```

```python
df_merged = pd.merge(df_chem, df_form[['LABNO', 'FORMATION']], on='LABNO')

```

```python
from sklearn.preprocessing import LabelEncoder

selected_icp40 = [  # your 35 selected elements
    'AL_ICP40', 'CA_ICP40', 'FE_ICP40', 'K_ICP40', 'NA_ICP40', 'TI_ICP40', 'AG_ICP40', 'CO_ICP40', 'CE_ICP40',
    'CD_ICP40', 'BI_ICP40', 'BE_ICP40', 'LI_ICP40', 'LA_ICP40', 'MO_ICP40', 'TH_ICP40', 'SN_ICP40', 'CR_ICP40',
    'CU_ICP40', 'GA_ICP40', 'PB_ICP40', 'SC_ICP40', 'NI_ICP40', 'Y_ICP40', 'V_ICP40', 'U_ICP40', 'ZN_ICP40',
    'BA_ICP40', 'MG_ICP40', 'SR_ICP40', 'MN_ICP40', 'P_ICP40', 'AU_ICP40', 'AS_ICP40', 'NB_ICP40'
]

df_model = df_merged.dropna(subset=selected_icp40 + ['FORMATION'])

# Encode formation
le = LabelEncoder()
df_model['formation_label'] = le.fit_transform(df_model['FORMATION'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

X = df_model[selected_icp40]
y = df_model['formation_label']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

y_pred = clf.predict(X_test)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(12, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted Formation")
plt.ylabel("Actual Formation")
plt.title("Confusion Matrix: Predicted vs Actual Formations")
plt.show()


```

```python
from sklearn.metrics import classification_report
import numpy as np

# Get the actual class labels used in this test split
labels_in_test = np.unique(y_test)
target_names = le.inverse_transform(labels_in_test)

print(classification_report(y_test, y_pred, labels=labels_in_test, target_names=target_names))
```
